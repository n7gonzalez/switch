# West Coast Wave Energy Capacity Factors from Synthetic Wave Data

Importing Packages

```python
# Third-party packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import h5pyd
import rex
import scipy
from scipy import interpolate
from scipy.interpolate import RegularGridInterpolator
import time
```

Importing power matrix

```python
#path should point to where the power matrix csv file is stored in your computer
#the power matrix can be downloaded from https://maps.nrel.gov/marine-energy-atlas/ under the capacity factor tab

path='/Users/nataliagonzalez/Documents/UCSD PhD/2022/Research 2022/Wave Energy/RM6_Power_Matrix_SAM_CSV.csv'
p_matrix = pd.read_csv(path, index_col=0)
p_matrix
```

Defining lookup axes and rated power of power matrix

```python
x = p_matrix.index.astype(float)
y = p_matrix.columns.astype(float)
z = p_matrix.values
rated_p = z.max() #rated power is max power in the power matrix
xp, yp = np.meshgrid(x,y,sparse=True) #makes a coordinate matrix relating the x and y parameters
```

2D interpolation function for imported power matrix

```python
#function for 2D interpolation of a coordinate matrix/grid (on orginal power matrix)

l = RegularGridInterpolator((x, y), z, bounds_error=False)
```

Extending power matrix

```python
max_Hs = 12.75 #input the max significant wave height that you want your extended power matrix to have
max_Te = 30.5 #input the max peak period that you want your extended power matrix to have

#changes step size of power matrix to 0.25
ext_x = np.arange(0, max_Hs, 0.25) 
ext_y = np.arange(0, max_Te, 0.25)

xp, yp = np.meshgrid(ext_x,ext_y, sparse=True) #makes a coordinate matrix relating the extended x and y parameters
ext_z = l((xp,yp)) #defines the values of the extended power matrix as the 2D interpolated values from the original matrix

#creates a dataframe for the extended power matrix
ext_p_matrix = pd.DataFrame(ext_z, index=ext_y, columns=ext_x)
ext_p_matrix = ext_p_matrix.T.fillna(method='ffill').fillna(method='bfill').fillna(0)
ext_p_matrix
```

```python
#creates a heat map of capacity factors associated with power values within the extended power matrix
plt.imshow(ext_p_matrix/rated_p, aspect = 'auto', origin='lower', extent=[min(ext_y),max(ext_y),min(ext_x),max(ext_x)])
cb = plt.colorbar()

#setting labels and font sizing
cb.set_label(label='Capacity Factor', size=16)
cb.ax.tick_params(labelsize=14)
plt.tick_params(labelsize=14)
plt.xlabel('$T_e\: (s)$', fontsize=16)
plt.ylabel('$H_s\: (m)$', fontsize=16)

#saving plot as an svg figure (vectorized image)
plt.savefig('p_matrix_heat_map.svg', format='svg', dpi=1400, bbox_inches='tight')
```

2D interpolation function for extended power matrix

```python
#function for 2D interpolation of a coordinate matrix/grid (on extended power matrix)
look = RegularGridInterpolator((ext_x, ext_y), ext_p_matrix.values, bounds_error=False)
```

Loading coordinates of dataset

```python
#path should point to where you saved the coordinates file on your computer generated from coordinates_in_dataset.ipynb
path = '/Users/nataliagonzalez/Documents/UCSD PhD/2022/Research 2022/Wave Energy/syn_loc_coordinates.csv'
coordinates = pd.read_csv(path, index_col=None)
coordinates
```

Importing wave data from NREL API and calculating capacity factors

```python
#set the granularity of data extraction desired
#for example, setting step to 1000 will give you every 1000 points in the dataset

ds = coordinates['loc_id'].max()  #endpoint of your array is the max value of loc_id in coordinates 
step = 1000

sparse_array=pd.Series(np.arange(0,ds,step))
sparse_array
```

```python
#this block extracts the wave characteristics from the ids you specified in sparse_array above

variables = ['significant_wave_height', 'peak_period'] #these are the wave characteristics you need to determine capactiy factors
pathAPI = '/nrel/US_wave/West_Coast/West_Coast_wave_2006.h5' #path in API that points to dataset you want
df_list = []

for idx in sparse_array:
    with h5pyd.File(pathAPI, mode='r') as f:
        # Extract time_index and convert to datetime
        time_index = pd.to_datetime(f['time_index'][...].astype(str))
        # Initialize DataFrame to store time-series data
        time_series = pd.DataFrame(index=[time_index])
        # Extract wave height and period
        new_cols = []
        for var in variables:
            # Get dataset
            ds = f[var]
            # Extract site idx and add to DataFrame
            time_series[var] = ds[:, idx]
            new_name = var + '_' + ds.attrs['units']
            new_cols.append(new_name)
    #creates time series with capacity factor and location id for each row of dataset
    time_series.columns = new_cols
    time_series.loc[:, 'capacity_factor'] = time_series.apply(lambda x: look((x['significant_wave_height_m'],x['peak_period_s']))/rated_p,axis=1)
    time_series.loc[:, 'loc_id'] = idx
    df_list.append(time_series)
    time.sleep(2) #adhearing to API download restrictions
df = pd.concat(df_list, ignore_index=False)
wave_data = df
```
